```python
import vk_api
```

```python
access_token = ""
```

```python
def get_vk_data(vk_user_id):
    vk_session = vk_api.VkApi(token=access_token)
    vk = vk_session.get_api()

    # –ü–æ–ª—É—á–µ–Ω–∏–µ –≥—Ä—É–ø–ø –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    groups = vk.groups.get(user_id=vk_user_id, extended=1)
    
    # –ü–æ–ª—É—á–µ–Ω–∏–µ –∑–∞–ø–∏—Å–µ–π —Å–æ —Å—Ç–µ–Ω—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    wall = vk.wall.get(owner_id=vk_user_id, count=100)

    # –ü–æ–ª—É—á–µ–Ω–∏–µ –ª–∞–π–∫–æ–≤ –¥–ª—è –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –ø–æ—Å—Ç–æ–≤ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 100)
    likes = []
    for item in wall['items']:
        post_likes = vk.likes.getList(type='post', owner_id=vk_user_id, item_id=item['id'], count=100)
        likes.append(post_likes)

    return groups, wall, likes
```

```python
def get_group_data(group_id):
    vk_session = vk_api.VkApi(token=access_token)
    vk = vk_session.get_api()

    # –ü–æ–ª—É—á–µ–Ω–∏–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –≥—Ä—É–ø–ø–µ
    group_info = vk.groups.getById(group_id=group_id, fields='members_count,description')

    # –ü–æ–ª—É—á–µ–Ω–∏–µ –∑–∞–ø–∏—Å–µ–π –∏–∑ –≥—Ä—É–ø–ø—ã
    group_posts = vk.wall.get(owner_id=-int(group_id), count=100)  # –ü—Ä–µ—Ñ–∏–∫—Å "-" –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è –≥—Ä—É–ø–ø—ã

    return group_info, group_posts
```

```python
def format_vk_output(group_info, group_posts):
    output = ""

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∏ —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –≥—Ä—É–ø–ø–µ
    if 'response' in group_info:
        for group in group_info['response']:
            output += f"–ù–∞–∑–≤–∞–Ω–∏–µ –≥—Ä—É–ø–ø—ã: {group['name']}\n"
            output += f"–û–ø–∏—Å–∞–Ω–∏–µ: {group['description']}\n"
            output += f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤: {group['members_count']}\n"
            output += f"–°—Å—ã–ª–∫–∞: https://vk.com/{group['screen_name']}\n"
            output += f"–§–æ—Ç–æ –≥—Ä—É–ø–ø—ã: {group['photo_200']}\n\n"

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∏ —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∑–∞–ø–∏—Å–µ–π –≥—Ä—É–ø–ø—ã
    if 'items' in group_posts:
        for post in group_posts['items']:
            output += f"–î–∞—Ç–∞ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏: {post['date']}\n"
            output += f"–¢–µ–∫—Å—Ç –ø–æ—Å—Ç–∞: {post.get('text', '–ë–µ–∑ —Ç–µ–∫—Å—Ç–∞')}\n"
            output += f"–õ–∞–π–∫–æ–≤: {post['likes']['count']}, –†–µ–ø–æ—Å—Ç–æ–≤: {post['reposts']['count']}\n"
            if 'views' in post:
                output += f"–ü—Ä–æ—Å–º–æ—Ç—Ä–æ–≤: {post['views']['count']}\n"
            output += "\n"

    return output
```

```python
group_info, group_posts = get_group_data('113716445')
```

```python
print(format_vk_output(group_info, group_posts))
```
